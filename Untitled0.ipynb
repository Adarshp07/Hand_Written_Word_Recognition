```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

from torchvision import datasets, transforms
```

```python
BATCH_SIZE = 64

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_ds = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_ds  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False)

print("Train:", len(train_ds), "Test:", len(test_ds))
```

```python
import torch
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)
model = MNISTCNN().to(DEVICE)
```

```python
class MNISTCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),  # 28x28
            nn.ReLU(),
            nn.MaxPool2d(2),                 # 14x14

            nn.Conv2d(32, 64, 3, padding=1), # 14x14
            nn.ReLU(),
            nn.MaxPool2d(2)                  # 7x7
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Dropout(0.25),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

model = MNISTCNN().to(DEVICE)
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

def train_one_epoch():
    model.train()
    total_loss, correct, total = 0.0, 0, 0

    for x, y in train_loader:
        x, y = x.to(DEVICE), y.to(DEVICE)

        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * x.size(0)
        correct += (logits.argmax(1) == y).sum().item()
        total += y.size(0)

    return total_loss/total, correct/total

@torch.no_grad()
def evaluate():
    model.eval()
    total_loss, correct, total = 0.0, 0, 0

    for x, y in test_loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        logits = model(x)
        loss = criterion(logits, y)

        total_loss += loss.item() * x.size(0)
        correct += (logits.argmax(1) == y).sum().item()
        total += y.size(0)

    return total_loss/total, correct/total
```

```python
EPOCHS = 5

train_losses, train_accs = [], []
test_losses, test_accs = [], []

for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_acc = train_one_epoch()
    te_loss, te_acc = evaluate()

    train_losses.append(tr_loss); train_accs.append(tr_acc)
    test_losses.append(te_loss); test_accs.append(te_acc)

    print(f"Epoch {epoch}: "
          f"train_loss={tr_loss:.4f}, train_acc={tr_acc:.4f} | "
          f"test_loss={te_loss:.4f}, test_acc={te_acc:.4f}")

plt.figure()
plt.plot(train_accs, label="train_acc")
plt.plot(test_accs, label="test_acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Accuracy vs Epoch")
plt.show()
```

```python
@torch.no_grad()
def get_all_preds():
    model.eval()
    all_preds, all_true = [], []
    for x, y in test_loader:
        x = x.to(DEVICE)
        logits = model(x).cpu()
        all_preds.append(logits.argmax(1))
        all_true.append(y)
    return torch.cat(all_true), torch.cat(all_preds)

y_true, y_pred = get_all_preds()

cm = torch.zeros(10, 10, dtype=torch.int32)
for t, p in zip(y_true, y_pred):
    cm[t, p] += 1

plt.figure(figsize=(6, 6))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
plt.xticks(range(10))
plt.yticks(range(10))
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()
from PIL import Image, ImageOps

preprocess = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((28, 28)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

def predict_image(path):
    img = Image.open(path).convert("L")

    # invert if background is bright (common case: white paper)
    if np.array(img).mean() > 127:
        img = ImageOps.invert(img)

    x = preprocess(img).unsqueeze(0).to(DEVICE)

    model.eval()
    with torch.no_grad():
        logits = model(x)
        probs = torch.softmax(logits, dim=1).squeeze().cpu().numpy()
        pred = int(np.argmax(probs))

    return pred, probs, img

filename = list(uploaded.keys())[0]
pred, probs, img = predict_image(filename)

print("Predicted digit:", pred)
print("Probabilities:", np.round(probs, 3))

plt.figure()
plt.imshow(img, cmap="gray")
plt.title(f"Prediction: {pred}")
plt.axis("off")
plt.show()

```
